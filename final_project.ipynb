–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–≥–æ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏—è –∑–∞–ø–∏—Å–µ–π —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ *'income_type'*:

````python
def procent_part(data, income_type):
    nan_count = data.loc[(data['days_employed'].isnull() == True)&(data['income_type'] == income_type), 'income_type'].count()
    all_count = data.loc[data['income_type'] == income_type, 'income_type'].count()
    procent = nan_count/all_count
    return '–ü—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ "' + income_type + '" {:.1%}'.format(procent)    
````

–ü—Ä–∏–≤–µ–¥–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏–∑ —Å–ø–∏—Å–∫–∞ *income_type_categories*: 

````python
for income_type in income_type_categories:
    print(procent_part(data, income_type))
````

<div class="alert alert-success">

–í–µ—Ä–Ω–æ, –∏–ª–∏ –º–æ–∂–Ω–æ –±—ã–ª–æ –æ—Å—Ç–∞–≤–∏—Ç—å –∫–∞–∫ –µ—Å—Ç—å

````python
data['days_employed'] = abs(data['days_employed'].astype('int'))
````

–ü—Ä–æ–≤–µ—Ä–∏–º, —Å–æ–¥–µ—Ä–∂–∏—Ç –ª–∏ —Å—Ç–æ–ª–±–µ—Ü *'total_income'* –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –ù–∞–π–¥–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —ç—Ç–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞:

````python
data['total_income'].min()
````

````python
data['total_income'].max()
````

–ö–∞–∫ –≤–∏–¥–∏–º, –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ –Ω–µ—Ç.

–í —Å–≤—è–∑–∏ —Å —Ç–µ–º, —Ç–æ –æ–±—ä–µ–º –¥–∞–Ω–Ω—ã—Ö —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ —Å—Ç–æ–ª–±—Ü—É *'total_income'* –¥–æ–≤–æ–ª—å–Ω–æ –±–æ–ª—å—à–æ–π –∏ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø–æ—á—Ç–∏ —É –≤—Å–µ—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ —Ç–∏–ø—É –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ (—Å—Ç–æ–ª–±–µ—Ü *'income_type'*), –º—ã –Ω–µ –º–æ–∂–µ–º —É–¥–∞–ª—è—Ç—å –∏–ª–∏ –∑–∞–º–µ—â–∞—Ç—å –Ω—É–ª—è–º–∏ —ç—Ç–∏ –ø—Ä–æ–ø—É—Å–∫–∏, —Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ –º–æ–∂–µ—Ç —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –æ—Ç—Ä–∞–∑–∏—Ç—Å—è –Ω–∞ –¥–∞–ª—å–Ω–µ–π—à–∏—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞—Ö –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è (–≤–æ–ø—Ä–æ—Å –æ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∫—Ä–µ–¥–∏—Ç–Ω–æ–π –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç–∏ –∫–ª–∏–µ–Ω—Ç–∞ –æ—Ç –µ–≥–æ –¥–æ—Ö–æ–¥–∞). –ü–æ—ç—Ç–æ–º—É –ø—Ä–∞–≤–∏–ª—å–Ω–µ–µ –±—É–¥–µ—Ç –∑–∞–º–µ–Ω–∏—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ *'total_income'* –Ω–∞ –µ–≥–æ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∫–∞–∂–¥–æ–π –∏–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ *income_type_categories*. 
–°–æ–∑–¥–∞–¥–∏–º –º–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç —Å—á–∏—Ç–∞—Ç—å –º–µ–¥–∏–∞–Ω—É –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏ –∑–∞–ø–æ–ª–Ω—è—Ç—å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ *'total_income'* –¥–ª—è —ç—Ç–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤. –ú–µ—Ç–æ–¥ —Ç–∞–∫–∂–µ –∑–∞–ø–æ–ª–Ω–∏—Ç —Å–ø–∏—Å–æ–∫ –∏–∑ –∑–Ω–∞—á–µ–Ω–∏–π –º–µ–¥–∏–∞–Ω—ã –¥–ª—è –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤:

````python
def income_type_mediana(data, income_type, median_list):
    data_income_type = data.loc[data['income_type'] == income_type]
    median = data_income_type['total_income'].median()
    data.loc[data['income_type'] == income_type, 'total_income'] = data.loc[data['income_type'] == income_type, 'total_income'].fillna(median)
    median_list.append(median)
````

–ü—Ä–∏–º–µ–Ω–∏–º –µ–≥–æ –∫ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞ *income_type_categories* –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–∏–µ –º–µ–¥–∏–∞–Ω—ã —É –Ω–∞—Å –ø–æ–ª—É—á–∏–ª–∏—Å—å –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å–ø–∏—Å–∫–∞: 

````python
median_list = []
for income_type in income_type_categories:
    income_type_mediana(data, income_type, median_list)
print(median_list)    
````

–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–≤ —ç—Ç–æ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –º—ã –º–æ–∂–µ–º –≤—ã–¥–µ–ª–∏—Ç—å 5 –æ—Å–Ω–æ–≤–Ω—ã—Ö –ª–µ–º–º: *'–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å', '–∞–≤—Ç–æ–º–æ–±–∏–ª—å', '—Å–≤–∞–¥—å–±–∞', '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ', '—Ä–µ–º–æ–Ω—Ç'*. –í—Å–µ –ª–µ–º–º—ã —Ç–∏–ø–∞ *'–∂–∏–ª—å–µ'* –∏ *'–∂–∏–ª–æ–π'* –æ—Ç–Ω–µ—Å–µ–º –∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ *'–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å'*. –ï—Å–ª–∏ –±—ã –≤ —Å—Ç–æ–ª–±—Ü–µ *'purpose'* –Ω–µ –±—ã–ª–æ —Ç–∞–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –∫–∞–∫ *'–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å'* –∏ *'–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'*, —Ç–æ –º—ã –±—ã –º–æ–≥–ª–∏ —Ä–∞—Å—à–∏—Ä–∏—Ç—å —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–π, —É–∫–∞–∑–∞–≤ –µ—â–µ –æ—Ç–¥–µ–ª—å–Ω–æ —Ç–∏–ø—ã –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏ –∏ –≤–∏–¥ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è, –Ω–æ —Ç–∞–∫–∏–µ –Ω–µ–æ–¥–Ω–æ–∑–Ω–∞—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç, –∑–Ω–∞—á–∏—Ç —Å–¥–µ–ª–∞—Ç—å –º—ã —ç—Ç–æ–≥–æ –Ω–µ –º–æ–∂–µ–º. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å –∫–∞—Ç–µ–≥–æ—Ä–∏–π: ***{'–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å', '–∞–≤—Ç–æ–º–æ–±–∏–ª—å', '—Å–≤–∞–¥—å–±–∞', '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ', '—Ä–µ–º–æ–Ω—Ç'}***.
–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏—é –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ *'purpose'*. –î–ª—è –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ —É–¥–æ–±–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±–∏–±–ª–∏–æ—Ç–µ–∫—É *PyMystem*:

````python
def purpose_group(purpose):
    lemma_purpose = m.lemmatize(purpose)
    if ('–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å' in lemma_purpose) or ('–∂–∏–ª—å–µ' in lemma_purpose) and ('—Ä–µ–º–æ–Ω—Ç' not in lemma_purpose):
        return '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å'
    elif '–∞–≤—Ç–æ–º–æ–±–∏–ª—å' in lemma_purpose:
        return '–∞–≤—Ç–æ–º–æ–±–∏–ª—å'
    elif '—Å–≤–∞–¥—å–±–∞' in lemma_purpose:
        return '—Å–≤–∞–¥—å–±–∞'
    elif '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ' in lemma_purpose:
        return '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'
    elif '—Ä–µ–º–æ–Ω—Ç' in lemma_purpose:
        return '—Ä–µ–º–æ–Ω—Ç'
````

–ñ–∏—Ä–Ω—ã–π –ø–ª—é—Å! –≠—Ç–æ –æ–¥–∏–Ω –∏–∑ —Å–ª–æ–∂–Ω–µ–π—à–∏—Ö —ç—Ç–∞–ø–æ–≤ –ø—Ä–æ–µ–∫—Ç–∞, —Å–ø—Ä–∞–≤–∏–ª–∞—Å—å –Ω–∞ –æ—Ç–ª–∏—á–Ω–æ


–ü—Ä–æ–≤–µ—Ä–∏–º, –∫–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è, –ø–µ—Ä–µ–±—Ä–∞–≤ –∑–Ω–∞—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ *data_purpose*:

````python
for purpose in data_purpose:
    print(purpose_group(purpose))
````

–ù–∞—à–∞ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞–±–æ—Ç–∞–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ, –º–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å –µ–µ –∫ –¥–∞–Ω–Ω—ã–º —Å—Ç–æ–ª–±—Ü–∞ *'purpose'* —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ apply(). 

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

````python
data['purpose_group'] = data['purpose'].apply(purpose_group)
data.head(10)
````

–ü—Ä–µ–¥–ª–∞–≥–∞—é —Ç–∞–∫–æ–π –≤–∞—Ä–∏–∞–Ω—Ç:

—Å—Ç–æ–ª–±–µ—Ü 'quantity_total' —Å –æ–±—â–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏;
—Å—Ç–æ–ª–±–µ—Ü 'quantity_of_debts' —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –¥–æ–ª–∂–Ω–∏–∫–æ–≤ –ø–æ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏;
—Å—Ç–æ–ª–±–µ—Ü 'percent_of_debts' —Å –¥–æ–ª–µ–π –¥–æ–ª–∂–Ω–∏–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö:</span>

````python
def ratio_result(data, categories):
    data_result = pd.DataFrame()
    data_result['quantity_total'] = data.groupby(categories)[categories].count()
    data_result['quantity_of_debts'] = data.groupby(categories)['debt'].sum()
    data_result['percent_of_debts'] = round(data_result['quantity_of_debts']/data_result['quantity_total']*100,1)
    return data_result
````
