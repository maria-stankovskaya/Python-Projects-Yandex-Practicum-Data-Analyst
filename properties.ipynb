–û—Å—Ç–∞–ª—å–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–∏ –Ω–µ —É—á–∞—Å—Ç–≤—É—é—Ç, –ø–æ—ç—Ç–æ–º—É –ø—Ä–æ–≤–µ—Ä—è—Ç—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏–π –∏ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –≤ –Ω–∏—Ö –º—ã –Ω–µ –±—É–¥–µ–º.

### –í—ã–≤–æ–¥

–ò–∑—É—á–µ–Ω—ã –¥–∞–Ω–Ω—ã–µ, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã–µ –≤ —Ñ–∞–π–ª–µ. –û–ø—Ä–µ–¥–µ–ª–µ–Ω—ã —Å—Ç—Ä–∞—Ç–µ–≥–∏—á–µ—Å–∫–∏ –≤–∞–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–æ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Ç–∞–º, –≥–¥–µ —ç—Ç–æ —É–º–µ—Å—Ç–Ω–æ. –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∫–∞ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –≤ –Ω—É–∂–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö —Å –¥–∞—Ç–∞–º–∏. –í—ã–ø–æ–ª–Ω–µ–Ω–∞ –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤.

## –†–∞—Å—á—ë—Ç—ã –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü—É

–°—Ç–æ–ª–±–µ—Ü —Å —Ü–µ–Ω–æ–π –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –±—ã–ª —Ä–∞—Å—Å—á–∏—Ç–∞–Ω –≤—ã—à–µ:

–î–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å —É–∫–∞–∑–∞–Ω–∏–µ–º –¥–Ω—è –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü–∞ –∏ –≥–æ–¥–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è:

````python
data['weekday_of_publication'] = data['first_day_exposition'].dt.weekday
data['month_of_publication'] = data['first_day_exposition'].dt.month
data['year_of_publication'] = data['first_day_exposition'].dt.year
data.head()
````

–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å —É–∫–∞–∑–∞–Ω–∏–µ —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã, –≤—ã–±–∏—Ä–∞—è –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ 3 –∫–∞—Ç–µ–≥–æ—Ä–∏–π - {'–ø–µ—Ä–≤—ã–π', '–ø–æ—Å–ª–µ–¥–Ω–∏–π', '–¥—Ä—É–≥–æ–π'}:

````python
def get_category(floor, floors_total):
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    elif floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    else:
        return '–¥—Ä—É–≥–æ–π'

    
data['floor_category'] = data.apply(lambda x: get_category(x['floor'], x['floors_total']), axis = 1)
data.head()
````
